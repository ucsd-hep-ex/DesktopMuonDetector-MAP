```python
import random;
import numpy as np;
import matplotlib.pyplot as plt;

file1 = open("myFile.txt", "w" )

for i in range (100):
    count = random.randint(0,10);
    time_stamp = random.randint(0,60);
    adc = random.randint(0,1023);
    voltage = random.randint(0,5);
    measurement_deadtime = random.randint(0,10);
    temperatureC = random.randint(0,40);

    file1.write(str(count) + " " + str(time_stamp) + " " + str(adc) + " " + str(voltage) + " " + str(measurement_deadtime) + " " + str(temperatureC) + "\n") 
#get_sipm_voltage becomes just voltage bc the function in the original code is for converting from adc
              
file1.close()

file1 = open("myFile.txt", "r" )

fileList = [];
for line in file1:
    line = line.strip();
    line = line.split(" ");
    lineList = [];
    for var in line:
        variable = int(var)
        lineList.append(variable)
    fileList.append(lineList)
    
countList = [];
time_stampList = [];
adcList = [];
voltageList = [];
measurement_deadtimeList = [];
temperatureCList = [];

for num in fileList:
    count = num[0];
    countList.append(count);
    time_stamp = num[1];
    time_stampList.append(time_stamp);
    adc = num[2];
    adcList.append(adc);
    voltage = num[3];
    voltageList.append(voltage);
    measurement_deadtime = num[4];
    measurement_deadtimeList.append(measurement_deadtime);
    temperatureC = num[5];
    temperatureCList.append(temperatureC);

file1.close()

plt.hist(countList)
plt.title("Muon Count Histogram")
plt.xlim(0, 10)
plt.ylim(0, 30)
plt.show()

plt.hist(time_stampList)
plt.title("Time Stamp Histogram")
plt.xlim(0, 60)
plt.ylim(0, 20)
plt.show()

plt.hist(adcList)
plt.title("ADC Value Histogram")
plt.xlim(0, 1023)
plt.ylim(0, 20)
plt.show()

plt.hist(voltageList)
plt.title("Voltage Histogram")
plt.xlim(0, 5)
plt.ylim(0, 40)
plt.show()

plt.hist(measurement_deadtimeList)
plt.title("Measurement Deadtime Histogram")
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.show()

plt.hist(temperatureCList)
plt.title("Temperature in C Histogram")
plt.xlim(0, 40)
plt.ylim(0, 20)
plt.show()


```
